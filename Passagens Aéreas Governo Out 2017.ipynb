# Análise Gastos do Governo com Passagens Aéreas
## Base de dados do SCDP - Outubro/2017



### Importando os dados

```python
import pandas as pd
```

```python
# Carrega a base de dados
df = pd.read_excel('201710_EmissaoPassagens_SCDP.xlsx', header=3, skipfooter=3)
```

```python
# Verifica os primeiros registros
df.tail()
```

```python
# Verifica os últimos registros
df.tail()
```

```python
df.info()
```

```python
# Correção nos nome de colunas com espaço
df.rename(columns=lambda x: x.strip(), inplace=True)

# Correção nos valores com espaço
df['Cancelado'] = df['Cancelado'].str.strip()
df['Nome do órgão superior'] = df['Nome do órgão superior'].str.strip()


# Conversão de colunas datetime
df['Data Emissão Bilhete'] = pd.to_datetime(df['Data Emissão Bilhete'])
df['Data Embarque'] = pd.to_datetime(df['Data Embarque'])
```

```python
# Criação coluna dia da semana
dayOfWeek={0:'Seg', 1:'Ter', 2:'Qua', 3:'Qui', 4:'Sex', 5:'Sáb', 6:'Dom'}
df['Dia da Semana do Embarque'] = df['Data Embarque'].dt.dayofweek.map(dayOfWeek)

# Criação coluna com o turno que o vôo foi realizado
def turno(hora):
    if (hora >= 6 and hora < 12):
        return 'Manhã'
    elif (hora >= 12 and hora < 18):
        return 'Tarde'
    elif (hora >= 18 and hora <= 23):
        return 'Noite'
    else:
        return 'Madrugada'

df['Turno'] = df['Data Embarque'].dt.hour.apply(lambda x: turno(x))

df.head()
```

# Análise dos dados

## Vôos por dia da semana e horários

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Base de dados de vôos não cancelados
df2 = df[df['Cancelado'] == 'Não']
```

```python
sns.factorplot(x='Dia da Semana do Embarque', data=df2, kind='count', order=['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom'], size=6)
plt.title('Quantidade de Vôos por dia da semana')
```

```python
sns.factorplot(x='Dia da Semana do Embarque', data=df2, kind='count', col='Turno', size=6)
```

```python
df2['Data Embarque'].dt.day.value_counts().plot.bar(x='Dia do mês', y='Número de vôos')
plt.title('Número de vôos por dia do mês')
```

## Top 5 órgãos com maior número de vôos

```python
# Conta a quantidade de vôos para cada órgão do governo
qtde_voos_orgao = df['Nome do órgão superior'].value_counts()
top_5 = qtde_voos_orgao[1:6]

labels = top_5.axes[0].tolist()
sizes = top_5.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)

plt.tight_layout()
plt.show()
```

## Top 5 órgãos que mais gastaram com vôos

```python
top_5_valores = df.groupby('Nome do órgão superior')['Valor Bilhete'].agg(['sum'])
top_5_valores.sort_values(by='sum', ascending=False, inplace=True)
top_5_valores = top_5_valores[1:6]

top_5_valores.plot.bar()
```
